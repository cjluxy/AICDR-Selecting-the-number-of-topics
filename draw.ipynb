```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize = (14,6))
ax=ax.reshape(6)

plt.subplots_adjust(hspace=0.3,wspace=0.3)
x=list(range(2,21))
x1=x[:-1]
def y_new(y):
    y_new=[]
    for i in range(len(y)):
        if i<len(y)-1:
            rate=abs(y[i]-y[i+1])
            # rate=abs(y[i+1]-y[i])/y[i]
            y_new.append(rate)
    return y_new


#1

y1=y_new(data['bbcsport_lda'].tolist())
y2=y_new(data['bbcsport_nmf'].tolist())
y3=y_new(data['bbcsport_kmeans'].tolist())
ax[0].plot(x1,y1,color='dodgerblue')
ax[0].plot(x1,y2,color='orange')
ax[0].plot(x1,y3,color='greenyellow')
ax[0].legend(labels=['LDA','NMF','K-Means'], loc='upper right',prop={'size': 12})
ax[0].set_xlabel('(a) BBCsport',size=10)
ax[0].set_ylabel('AICDR',size=10)
ax[0].scatter(x=5,y=np.max(y1),s=50,color='dodgerblue')
ax[0].scatter(x=4,y=np.max(y2),s=50,color='orange')
ax[0].scatter(x=5,y=np.max(y3),s=50,color='greenyellow')
ax[0].axvline(x=4.5, color='gray', linestyle='--') 
ax[0].axvline(x=5.5, color='gray', linestyle='--') 


#2
y1=y_new(data['bbcnews_lda'].tolist())
y2=y_new(data['bbcnews_nmf'].tolist())
y3=y_new(data['bbcnews_kmeans'].tolist())
ax[1].plot(x1,y1,color='dodgerblue')
ax[1].plot(x1,y2,color='orange')
ax[1].plot(x1,y3,color='greenyellow')
ax[1].legend(labels=['LDA','NMF','K-Means'], loc='upper right',prop={'size': 12})
ax[1].set_xlabel('(b) BBCNews',size=10)
ax[1].set_ylabel('AICDR',size=10)
ax[1].scatter(x=6,y=np.max(y1),s=50,color='dodgerblue')
ax[1].scatter(x=5,y=np.max(y2),s=50,color='orange')
ax[1].scatter(x=6,y=np.max(y3),s=50,color='greenyellow')
ax[1].axvline(x=4.5, color='gray', linestyle='--') 
ax[1].axvline(x=5.5, color='gray', linestyle='--') 
#3
y1=y_new(data['reuter_lda'].tolist())
y2=y_new(data['reuter_nmf'].tolist())
y3=y_new(data['reuter_kmeans'].tolist())
ax[2].plot(x1,y1,color='dodgerblue')
ax[2].plot(x1,y2,color='orange')
ax[2].plot(x1,y3,color='greenyellow')
ax[2].legend(labels=['LDA','NMF','K-Means'], loc='upper right',prop={'size': 12})
ax[2].set_xlabel('(c) Reuters',size=10)
ax[2].set_ylabel('AICDR',size=10)
ax[2].scatter(x=3,y=np.max(y1),s=50,color='dodgerblue')
ax[2].scatter(x=3,y=np.max(y2),s=50,color='orange')
ax[2].scatter(x=3,y=np.max(y3),s=50,color='greenyellow')
ax[2].axvline(x=2.5, color='gray', linestyle='--') 
ax[2].axvline(x=3.5, color='gray', linestyle='--') 

#4
y1=y_new(data['agnews_gsdmm'].tolist())
y2=y_new(data['agnews_seanmf'].tolist())
ax[3].plot(x1,y1,color='dodgerblue')
ax[3].plot(x1,y2,color='orange')
ax[3].legend(labels=['GSDMM','SeaNMF'], loc='upper right',prop={'size': 12})
ax[3].set_xlabel('(d) AGNews',size=10)
ax[3].set_ylabel('AICDR',size=10)
ax[3].scatter(x=4,y=np.max(y1),s=50,color='dodgerblue')
ax[3].scatter(x=5,y=np.max(y2),s=50,color='orange')
ax[3].axvline(x=4.5, color='gray', linestyle='--') 
ax[3].axvline(x=3.5, color='gray', linestyle='--') 
#5
y1=y_new(data['snippets1_gsdmm'].tolist())
y2=y_new(data['snippets1_seanmf'].tolist())
ax[4].plot(x1,y1,color='dodgerblue')
ax[4].plot(x1,y2,color='orange')
ax[4].legend(labels=['GSDMM','SeaNMF'], loc='upper right',prop={'size': 12})
ax[4].set_xlabel('(e) Snippets-1',size=10)
ax[4].set_ylabel('AICDR',size=10)
ax[4].scatter(x=4,y=np.max(y1),s=50,color='dodgerblue')
ax[4].scatter(x=4,y=np.max(y2),s=50,color='orange')
ax[4].axvline(x=4.5, color='gray', linestyle='--') 
ax[4].axvline(x=3.5, color='gray', linestyle='--') 
#6
y1=y_new(data['snippets2_gsdmm'].tolist())
y2=y_new(data['snippets2_seanmf'].tolist())
ax[5].plot(x1,y1,color='dodgerblue')
ax[5].plot(x1,y2,color='orange')
ax[5].legend(labels=['GSDMM','SeaNMF'], loc='upper right',prop={'size': 12})
ax[5].set_xlabel('(f) Snippets-2',size=10)
ax[5].set_ylabel('AICDR',size=10)
ax[5].scatter(x=3,y=np.max(y1),s=50,color='dodgerblue')
ax[5].scatter(x=3,y=np.max(y2),s=50,color='orange')
ax[5].axvline(x=4.5, color='gray', linestyle='--') 
ax[5].axvline(x=3.5, color='gray', linestyle='--')     
# plt.savefig(r'.\output\AICDR.png',dpi=1000,bbox_inches = 'tight') 
```
